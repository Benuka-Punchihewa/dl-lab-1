```python
# Pandas for managing datasets
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('Au_nanoparticle_dataset.csv')

df.head()
```

```python
# Drop unwanted columns
new_df = df.filter(items = ["N_total", "N_bulk", "N_surface", "R_avg"], axis = 1)

new_df.head(20)
```

```python
mean = new_df['N_total'].mean()
standard_deviation = new_df['N_total'].std()
q1 = new_df['N_total'].quantile(0.25)
q2 = new_df['N_total'].quantile(0.50)
q3 = new_df['N_total'].quantile(0.75)

print(f"Mean of N_total: {mean}")
print(f"Standard deviation of N_total: {standard_deviation}")
print(f"First quartile of N_total: {q1}")
print(f"Second quartile of N_total: {q2}")
print(f"Third quartile of N_total: {q3}")
```

```python
mean = new_df['N_bulk'].mean()
standard_deviation = new_df['N_bulk'].std()
q1 = new_df['N_bulk'].quantile(0.25)
q2 = new_df['N_bulk'].quantile(0.50)
q3 = new_df['N_bulk'].quantile(0.75)

print(f"Mean of N_bulk: {mean}")
print(f"Standard deviation of N_bulk: {standard_deviation}")
print(f"First quartile of N_bulk: {q1}")
print(f"Second quartile of N_bulk: {q2}")
print(f"Third quartile of N_bulk: {q3}")
```

```python
mean = new_df['N_surface'].mean()
standard_deviation = new_df['N_surface'].std()
q1 = new_df['N_surface'].quantile(0.25)
q2 = new_df['N_surface'].quantile(0.50)
q3 = new_df['N_surface'].quantile(0.75)

print(f"Mean of N_surface: {mean}")
print(f"Standard deviation of N_surface: {standard_deviation}")
print(f"First quartile of N_surface: {q1}")
print(f"Second quartile of N_surface: {q2}")
print(f"Third quartile of N_surface: {q3}")
```

```python
mean = new_df['R_avg'].mean()
standard_deviation = new_df['R_avg'].std()
q1 = new_df['R_avg'].quantile(0.25)
q2 = new_df['R_avg'].quantile(0.50)
q3 = new_df['R_avg'].quantile(0.75)

print(f"Mean of R_avg: {mean}")
print(f"Standard deviation of R_avg: {standard_deviation}")
print(f"First quartile of R_avg: {q1}")
print(f"Second quartile of R_avg: {q2}")
print(f"Third quartile of R_avg: {q3}")
```

```python
import matplotlib.pyplot as plt

# Create a 1x4 layout for histograms
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

# Plot histograms for each feature
df['N_total'].plot.hist(ax=axes[0], bins=10, edgecolor='black')
axes[0].set_title('N_total')

df['N_bulk'].plot.hist(ax=axes[1], bins=10, edgecolor='black')
axes[1].set_title('N_bulk')

df['N_surface'].plot.hist(ax=axes[2], bins=10, edgecolor='black')
axes[2].set_title('N_surface')

df['R_avg'].plot.hist(ax=axes[3], bins=10, edgecolor='black')
axes[3].set_title('R_avg')
```

```python
import seaborn as sns

# Create a pairplot
sns.pairplot(new_df)

# Show the plot
plt.show()
```

```python
g = sns.PairGrid(new_df)

# Diagonal plots with histogram and kde
g.map_diag(sns.histplot, kde=True)

# Upper triangle with bivariate histogram
g.map_upper(sns.histplot, bins=10, cbar=True)

# Lower triangle with bivariate kde
g.map_lower(sns.kdeplot, cmap='Blues_d')

# Show the plot
plt.show()
```
