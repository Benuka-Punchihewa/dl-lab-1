```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the scale parameter (lambda) for the exponential distribution
scale = 1

# Generate a random 4x4 array from an exponential distribution
random_array = np.random.exponential(scale, size=(4, 4))

print(random_array)
```

```python
# Generate a random 100000x1 array from an exponential distribution
random_exponential_array = np.random.exponential(scale, size=(100000, 1))

print(random_exponential_array)
```

```python
# Generate random samples from uniform and normal distributions for comparison
uniform_samples = np.random.uniform(0, scale, size=(100000,))
normal_samples = np.random.normal(scale, scale, size=(100000,))

plt.hist(random_exponential_array, density=True, bins=100, histtype="step", color="blue", label="Exponential distribution")
plt.hist(uniform_samples, density=True, bins=100, histtype="step", color="red", label="Uniform distribution")
plt.hist(normal_samples, density=True, bins=100, histtype="step", color="green", label="Normal distribution")
plt.axis([-1, 6, 0, 1.1])
plt.legend(loc = "upper right")
plt.title("Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute the corresponding values of Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface plot of Z = X^2 + Y^2')

# Show the plot
plt.show()
```

```python
# Pandas for managing datasets
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Drop unwanted columns
df.drop(['Name', "Type 1", "Type 2", "Total", "Stage", "Legendary"], axis=1, inplace=True)

df.head()
```

```python
import seaborn as sns

# Calculate Pearson correlation matrix
pearson_corr = df.corr(method='pearson')

# Calculate Spearman rank correlation matrix
spearman_corr = df.corr(method='spearman')

# Set up the figure
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[0])
axes[0].set_title('Pearson Correlation')

# Plot Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[1])
axes[1].set_title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
